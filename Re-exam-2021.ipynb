```python
t = symbols('t', positive=True)
y1, y2 = symbols('y1 y2', cls=Function)

eq1 = Eq(y1(t).diff(t), -(8/200)*y1(t) + (3/100)*y2(t))
eq2 = Eq(y2(t).diff(t), (8/200)*y1(t) - (8/100)*y2(t))

# Solve
solutions = dsolve([eq1, eq2], [y1(t), y2(t)], ics={y1(0): 100, y2(0): 10})
y1_sol = solutions[0].rhs
y2_sol = solutions[1].rhs

print(f"\nSolutions:")
print(f"y1(t) = {y1_sol}")
print(f"y2(t) = {y2_sol}")

# Calculate salt amounts at t = 11
t_target = 11
y1_at_11 = float(y1_sol.subs(t, t_target))
y2_at_11 = float(y2_sol.subs(t, t_target))

print(f"\nAt t = {t_target} minutes:")
print(f"Salt in tank A: {y1_at_11:.2f}g → {round(y1_at_11)}g")
print(f"Salt in tank B: {y2_at_11:.2f}g → {round(y2_at_11)}g")

print(f"\nFinal Answer:")
print(f"Tank A: {round(y1_at_11)} grams")
print(f"Tank B: {round(y2_at_11)} grams")
```
